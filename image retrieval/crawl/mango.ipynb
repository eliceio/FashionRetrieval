```python
## 웹 크롤링 코드
## MANGO 의류 사이트: https://shop.mango.com/kr

#################################### URL 목록 모두 생성 시작

from bs4 import BeautifulSoup
from selenium import webdriver
import numpy as np
import urllib.request
import time


# 드라이버 생성
# chromedriver 설치된 경로를 정확히 기재해야 함
# Chrome의 경우 | 아까 받은 chromedriver의 위치를 지정해준다.
path = 'C:\\Users\\Hyunjoo\\Documents\\chromedriver_win32\\chromedriver.exe'


# # 여성 / 셔츠
# url = "https://shop.mango.com/kr/%E1%84%8B%E1%85%A7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%89%E1%85%A7%E1%84%8E%E1%85%B3_c78920337"
# # 여성 / 티셔츠&탑 
# url = 'https://shop.mango.com/kr/%E1%84%8B%E1%85%A7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%90%E1%85%B5%E1%84%89%E1%85%A7%E1%84%8E%E1%85%B3---%E1%84%90%E1%85%A1%E1%86%B8_c66796663'
# 여성 / 스웨트셔츠  
url = "https://shop.mango.com/kr/%E1%84%8B%E1%85%A7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%89%E1%85%B3%E1%84%8B%E1%85%B0%E1%84%90%E1%85%B3%E1%84%89%E1%85%A7%E1%84%8E%E1%85%B3_c12454089"

# ## 여성 / 가디건 & 스웨터 
# url = "https://shop.mango.com/kr/%E1%84%8B%E1%85%A7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%80%E1%85%A1%E1%84%83%E1%85%B5%E1%84%80%E1%85%A5%E1%86%AB---%E1%84%89%E1%85%B3%E1%84%8B%E1%85%B0%E1%84%90%E1%85%A5_c18200786"
# ## 여성 / 재킷 
# url = "https://shop.mango.com/kr/%E1%84%8B%E1%85%A7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%8C%E1%85%A2%E1%84%8F%E1%85%B5%E1%86%BA_c69427016"


# ## 남성 / 티셔츠  
# url = "https://shop.mango.com/kr/%E1%84%82%E1%85%A1%E1%86%B7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%90%E1%85%B5%E1%84%89%E1%85%A7%E1%84%8E%E1%85%B3_c12018147"
# ## 남성 / 폴로셔츠  
# url = "https://shop.mango.com/kr/%E1%84%82%E1%85%A1%E1%86%B7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%91%E1%85%A9%E1%86%AF%E1%84%85%E1%85%A9-%E1%84%89%E1%85%A7%E1%84%8E%E1%85%B3_c25175143"
# ## 남성 / 스웨트셔츠  
# url = "https://shop.mango.com/kr/%E1%84%82%E1%85%A1%E1%86%B7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%89%E1%85%B3%E1%84%8B%E1%85%B0%E1%84%90%E1%85%B3%E1%84%89%E1%85%A7%E1%84%8E%E1%85%B3_c71156082"

# ## 남성 / 가디건
# url = "https://shop.mango.com/kr/%E1%84%82%E1%85%A1%E1%86%B7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%80%E1%85%A1%E1%84%83%E1%85%B5%E1%84%80%E1%85%A5%E1%86%AB_c33749244"
# ## 남성 / 재킷  
# url = "https://shop.mango.com/kr/%E1%84%82%E1%85%A1%E1%86%B7%E1%84%89%E1%85%A5%E1%86%BC/%E1%84%8C%E1%85%A2%E1%84%8F%E1%85%B5%E1%86%BA_c16042202"


driver = webdriver.Chrome(path)
driver.implicitly_wait(10) 
driver.get(url) 
driver.find_element_by_xpath("//div[@class='order-catalog']/nav[@class='nav-columns']/button[@data-columns='4']").click()
time.sleep(1)

last_height = driver.execute_script("return document.body.scrollHeight")
top_url = []
while True: 
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(0.5)
    
    
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    ## URL 목록 생성
    for i in soup.find_all("div", {"class":"product-list-item product-layout-4-columns-A1 product-list-item-js"}):
        if i.find("a", {"class":"product-list-link product-list-link-js"}).get("href") not in top_url:
            top_url.append(i.find("a", {"class":"product-list-link product-list-link-js"}).get("href"))
    
    
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
driver.close()    


## 확인
print(np.array(top_url))
print(len(top_url))  

#################################### URL 목록 모두 생성 끝



##################################### crawling 시작

from bs4 import BeautifulSoup
from selenium import webdriver
import numpy as np
import urllib.request
import time

f = open("prod_info.txt", "w", encoding="utf8")
g = open("prod_gubun.txt", "w", encoding="utf8")

path = 'C:\\Users\\Hyunjoo\\Documents\\chromedriver_win32\\chromedriver.exe'

gubun = []
num = 1
for row in range(len(top_url)): 
    url = top_url[row]

    driver = webdriver.Chrome(path)
    driver.implicitly_wait(10)
    
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    
    ## 중복 상품을 구분하기 위한 변수
    name = soup.find("h1", {"class":"product-name"}).text
    prod_code = soup.find("span",{"class":"product-reference"}).text.strip("상품 코드. ").strip()
    prod_gubun = name + prod_code
    
    if prod_gubun not in gubun:
        gubun.append(prod_gubun)
        g.write(prod_gubun+"\n")

        
        ## 상품 color 출력 (반복된 color는 품절 상품인듯)
        img_c = driver.find_elements_by_css_selector("div.color-container > img")
        temp_img_color = []
        for i in img_c:
            temp_img_color.append(i.get_attribute("alt"))
#         print(temp_img_color)

        ## 상품 color 중 품절된 것 제외  
        img_color = []
        img_color_ind = []
        i = 0
        j = 0

        if (len(temp_img_color) > 1):
            while i < len(temp_img_color): 
                if i < (len(temp_img_color)-1):
                    if (temp_img_color[i] != temp_img_color[i+1]) & (temp_img_color[i-1] != temp_img_color[i]):
                        img_color.append(temp_img_color[i])
                        img_color_ind.append(i-j)
                    else:
                        i += 1
                        j += 1
                else:
                    if temp_img_color[i-1] != temp_img_color[i]:
                        img_color.append(temp_img_color[i])
                        img_color_ind.append(i-j)
                    else:
                        i += 1
                i += 1
        else:
            img_color = temp_img_color
            img_color_ind = [0]
#             print(img_color_ind)
        
        if len(img_color) == 0:
            next
            
            
        ## 상품 id 출력
        temp_img_id = []
        img_i = driver.find_elements_by_css_selector("div.color-container")
        for i in img_i:
            temp_img_id.append(i.get_attribute("id"))
#         print(temp_img_id)
        temp_img_id = list(temp_img_id)


        for itr in img_color_ind:
            driver.find_element_by_xpath("//div[@class='colors-container']/div[@id=" + "'" + str(temp_img_id[itr]) + "'" + "]/img[@class='color-image']").click()  # 버튼 클릭!
            time.sleep(3)
            
            img_u = driver.find_elements_by_css_selector("div.image-btn > img")
            img_url = []
            for i in img_u:
                img_url.append(i.get_attribute("src"))
            img_url = list(img_url)

            save_path = "C:/Users/Hyunjoo/Desktop/DS Academy/DS 팀프로젝트/crawling/image_mango2/"
            urllib.request.urlretrieve(img_url[-1], save_path + "img" + str(num) + '.jpg')
   
         
            ## 가격, 소재, 성별, URL 저장
            prod_price = soup.find("span", {"class":"product-sale"}).text.strip("￦").strip()
            material_text = soup.find_all("p", {"class":"product-info-text"})[len(soup.find_all("p", {"class":"product-info-text"}))-2].text.strip("조성섬유:").strip().split(".")[0]
            prod_material = material_text.replace(",", "/")
            gender = "F"  # 남성제품일경우 M으로 변경
            prod_url = url
            prod_name = name
            index = num

            info = str(index) + "|" + str(gender) + "|" + str(prod_name) + "|" + str(prod_url) + "|" + str(prod_price) + "|" + str(prod_material) + "\n"
            f.write(info)
            
            ## 값 확인
            print(info)
            time.sleep(2)

            num += 1
        driver.close()
    else:
        driver.close()
f.close()
g.close()

##################################### crawling 끝



## 상품 정보 파일 불러오기
import csv
with open('C:/Users/Hyunjoo/Desktop/DS Academy/DS 팀프로젝트/crawling/prod_info.txt','r', encoding="utf8") as f:
    prod_info = list(csv.reader(f, delimiter='|'))

print(np.array(prod_info))

```
